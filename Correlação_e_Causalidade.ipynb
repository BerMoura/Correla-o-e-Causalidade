```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados fictcios:
np.random.seed(99)
meses = np.arange(1, 31)  # Meses fictcios
internacoes = np.random.uniform(100, 300, 30)  # Quantidade de internaes
musicas = (internacoes * 0.15 + np.random.uniform(1, 5, 30)).clip(1, 50)  # Ajustando para um mximo de 50 msicas

# Criar DataFrame
df = pd.DataFrame({'Perodo': meses, 'Internaes': internacoes, 'Msicas Lanadas': musicas})

# Grfico de disperso com linha de regresso
sns.lmplot(x='Internaes', y='Msicas Lanadas', data=df, aspect=2, line_kws={'color': 'red'})
plt.title('Relao entre Internaes e Msicas Lanadas pelo Manuel Gomes')
plt.xlabel('Quantidade de Internaes')
plt.ylabel('Msicas Lanadas')
plt.show()

# Clculo da correlao
correlacao = df[['Internaes', 'Msicas Lanadas']].corr()
print("Matriz de Correlao:\n", correlacao)
```
