```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados fictícios:
np.random.seed(99)
meses = np.arange(1, 31)  # Meses fictícios
internacoes = np.random.uniform(100, 300, 30)  # Quantidade de internações
musicas = (internacoes * 0.15 + np.random.uniform(1, 5, 30)).clip(1, 50)  # Ajustando para um máximo de 50 músicas

# Criar DataFrame
df = pd.DataFrame({'Período': meses, 'Internações': internacoes, 'Músicas Lançadas': musicas})

# Gráfico de dispersão com linha de regressão
sns.lmplot(x='Internações', y='Músicas Lançadas', data=df, aspect=2, line_kws={'color': 'red'})
plt.title('Relação entre Internações e Músicas Lançadas pelo Manuel Gomes')
plt.xlabel('Quantidade de Internações')
plt.ylabel('Músicas Lançadas')
plt.show()

# Cálculo da correlação
correlacao = df[['Internações', 'Músicas Lançadas']].corr()
print("Matriz de Correlação:\n", correlacao)
```
